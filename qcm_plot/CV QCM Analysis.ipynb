## Calculate mass change with Sauerbrey equation

The Sauerbrey equation is (source [Wikipedia](https://en.wikipedia.org/wiki/Sauerbrey_equation)):

$\Delta f = -\frac{2f_0^2}{A \sqrt{ \rho_q \mu_q } }\Delta m$

where:
* $f_0$ – Resonant frequency (Hz)
* $\Delta f$ – Frequency change (Hz)
* $\Delta m$ – Mass change (g)
* $A$ – Piezoelectrically active crystal area (Area between electrodes, cm<sup>2</sup>)
* $\rho_q$  – Density of quartz ($\rho_q$ = 2.648 g/cm<sup>3</sup>)
* $\mu_q$ – Shear modulus of quartz for AT-cut crystal ($\mu_q = 2.947 \times 10^{11}$ g/cm<sup>2</sup>)

According to International Crystal Manufacturing Co (ICMFG):
* $A$ = 0.205 cm$^2$
* $\rho$ = 2.648 g/cm$^3$
* $\mu$ = $2.947 \times 10^{11}$ g/(cm s$^2$)

Note that the mass change cannot be greater than two percent of the resonant frequency. That means we cannot change freq more than:

```python
0.02 * 10 * 10**6 #Hz
```

which seems to be a lot more than what I observe (within 100's of Hz). ICM documentation says that for a 0.14 ng mass change for a 8 Mhz crystal, the corresponding frequency change is 0.1 Hz. Let's check that:

```python
def solve_freq_sauerbreyeq(delta_mass, resonant_freq = 10.0e6): #Hz
    A = 0.205 #cm^2
    mu = 2.947 * 10**11 #g/(cm s^2)
    p = 2.648 #g/cm^3
    
    return (delta_mass * -2 * resonant_freq**2)/(A * math.sqrt(mu * p))
print '%f g' % solve_freq_sauerbreyeq(0.14e-9, resonant_freq = 8.0e6) #g
```

Yep, that works out.

```python
def solve_mass_sauerbreyeq(delta_freq, resonant_freq = 10.0e6): #Hz
    A = 0.205 #cm^2
    mu = 2.947 * 10**11 #g/(cm s^2)
    p = 2.648 #g/cm^3
    
    return (delta_freq * A * math.sqrt(mu * p))/(-2 * resonant_freq**2)
print '%e g' % solve_mass_sauerbreyeq(-0.1, resonant_freq = 8.0e6) #Hz
print '%f ng' % (solve_mass_sauerbreyeq(-0.1, resonant_freq = 8.0e6) * 10**9)
```

Need to account for the baseline mass that's already on the quartz:

```python
def solve_mass_sauerbreyeq_for_freq(freq, baseline_freq = None, resonant_freq = 10e6): #Hz
    baseline_mass = 0.0
    if baseline_freq is not None:
        baseline_mass = solve_mass_sauerbreyeq_for_freq(baseline_freq)
    return solve_mass_sauerbreyeq(freq - resonant_freq) - baseline_mass
print '%e g' % solve_mass_sauerbreyeq_for_freq(9.980417e6, baseline_freq = 9.980416e6)
```

```python
# We average a few initial points to establish the baseline frequency
baseline_freq = numpy.average(qcm_segs[0]['fs'][0:5]) * 10**6 #Hz
for seg in qcm_segs:
    seg['delta_mass'] = seg['fs'].apply(lambda x: solve_mass_sauerbreyeq_for_freq(x * 10**6, baseline_freq = baseline_freq))
```

## Add potential column to QCM data

We want to add a potential column to the QCM data such that it corresponds to the potentials in the CVs. The difficulty is that the CV changes direction of potential, and it's a pain to try to calculate exactly when the potential changes. Thus, we use the strategy of doing a linear regression on each CV segment (potential = m * time + b). Then we use this equation to calculate potential for a given QCM time.

```python
for cv_seg, qcm_seg in zip(cv_segs, qcm_segs):
    model = sm.ols('potential ~ time', data = cv_seg)
    results = model.fit()
    #results.params #contains the Intercept and time
    #print results.summary()
    qcm_seg['potential'] = results.params['time'] * qcm_seg['t'] + results.params['Intercept']
```
